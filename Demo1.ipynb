```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#generic plot command
plt.plot([8,6,3,12,7])
plt.show()
```

```python
#change mark type
plt.plot([8,6,3,12,7], 'o', color='magenta', linestyle = 'dashed')
plt.show()
```

```python
plt.plot([8,6,3,12,7], 'gs')
plt.show()
```

```python
# read in data with pandas and store as object
lures = pd.read_excel('LURES.xlsx', sheet_name = 'LURES')
lures
```

```python
#object class
type(lures)
```

```python
lures.dtypes
```

```python
QTY=lures['QUANTITY']
SALES = lures['SALES']
```

```python
type(QTY)
```

```python
plt.plot(QTY, SALES, 'o')
```

```python
plt.scatter(QTY, SALES)
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(QTY, SALES, marker='s', color='orange')
plt.xlabel('Quantity')
plt.ylabel('Sales')
plt.title('Scatterplot of Sales vs. Quantity')
plt.show()
```

```python
# format the markers of the scatterplot
# axis labels and tickers
plt.figure(figsize = (12, 6), facecolor = 'whitesmoke')
plt.scatter(lures['QUANTITY'], lures['SALES'],
           marker = 's', edgecolors = 'black',
           facecolors = 'aqua', linewidths = 2, s = 100)
plt.xlabel('Quantity', fontsize = 20)
plt.ylabel('Sales', fontsize = 20)
plt.tick_params(axis = 'both', labelsize = 'large')
plt.title('Scatterplot', fontsize = 24, loc='left',
         fontstyle = 'oblique', color = 'blue')
plt.grid(color = 'lightgray', linestyle = '--', linewidth = 1)
plt.show()
```

```python
# check style sheet available
plt.style.available
```

```python
plt.style.use('ggplot')
plt.figure(figsize = (12, 6))
plt.scatter(lures['QUANTITY'], lures['SALES'])
plt.xlabel('Quantity')
plt.ylabel('Sales')
plt.title('Scatterplot - ggplot2')
plt.show()
```

```python
plt.style.use('dark_background')
plt.figure(figsize = (12, 6))
plt.scatter(lures['QUANTITY'], lures['SALES'])
plt.xlabel('Quantity')
plt.ylabel('Sales')
plt.title('Scatterplot - dark_background')
plt.show()
```
